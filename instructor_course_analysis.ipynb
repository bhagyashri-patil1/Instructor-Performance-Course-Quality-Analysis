```python
def rating_tier(rating):
    
    if rating >= 4.5:
        return "High Rated"
    
    elif rating >= 3.5:
        return "Mid Rated"
    
    else:
        return "Low Rated"

df["RatingTier"] = df["TeacherRating"].apply(rating_tier)
```

```python
df["RatingTier"].value_counts()
```

```python
enrollment_by_tier = df.groupby("RatingTier")["TransactionID"].count()

print(enrollment_by_tier)
```

```python
df.head(5)
```

```python
enrollment_by_tier.plot(kind="bar")
plt.title("Enrollments by Instructor Rating Tier")
plt.ylabel("Total Enrollments")
plt.show()
```
